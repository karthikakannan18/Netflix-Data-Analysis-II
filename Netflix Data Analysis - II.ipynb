## 2. Exploratory Analysis and Visualization
The data was analysed using data visulaisation to determine information about the different types od shows available on Netflix.

Analysis and visualisation libraries from seaborn and matplotlib were applied to the data as shown.

```python
import seaborn as sns
import matplotlib
import matplotlib.pyplot as plt
%matplotlib inline

sns.set_style('darkgrid')
matplotlib.rcParams['font.size'] = 14
matplotlib.rcParams['figure.figsize'] = (9, 5)
matplotlib.rcParams['figure.facecolor'] = '#00000000'
```

```python
# Analyzing the differences in type of shows
sns.countplot(x='type', data=data)
plt.title("Type of the Show on Netflix");
```

```python
# Top 10 Genres of Movies
plt.figure(figsize=(12,6))
data[data["type"]=="Movie"]["listed_in"].value_counts()[:10].plot(kind="barh",color="purple")
plt.title("Top 10 Genres of Movies",size=18);
```

```python
# Top 10 Genres of TV Shows
plt.figure(figsize=(12,6))
data[data["type"]=="TV Show"]["listed_in"].value_counts()[:10].plot(kind="barh",color="pink")
plt.title("Top 10 Genres of TV Shows",size=18);
```

```python
# Number of shows released per year
data = data[data['release_year']>2010]
sns.kdeplot(data=data['release_year'], label='release_year', shade=True)
plt.title('Number of shows released per year')
plt.show()
```

```python
#Types of shows by ratings
plt.figure(figsize=(11,11))
sns.countplot(x='rating',data=data)
plt.title("Ratings");
```

```python
# Movies vs TV Shows rating
plt.figure(figsize=(12,12))
sns.countplot(x='type',hue='rating',data=data)
plt.title("Movies vs Tv shows Rating");
```

```python
#Frequency of release over the last ten years
sns.countplot(data=data,x='release_year',hue='type');
```

```python
# Countries with maximum content creation
plt.figure(1, figsize=(15, 7))
plt.title("Country with maximum content creation")
sns.countplot(x = "country", order=data['country'].value_counts().index[0:10] ,data=data,palette='Accent');
```

```python
#Type of content produced by country

data_country = data.groupby(['country']).count()['show_id'].to_frame().reset_index()
data_country = data_country.sort_values('show_id',ascending=False)
data_country = data_country.head(5).reset_index(drop=True)
list_c = data_country['country']
top_8_data = data[data['country'].isin(list_c)]
year_with_show = top_8_data.groupby(['country','type']).count()['show_id'].to_frame().reset_index()
year_with_show = year_with_show.sort_values(by = 'show_id', ascending= False)
```

```python
plt.style.use('seaborn-pastel')
plt.figure(figsize=(20,10))
sns.barplot(x="country", y="show_id", hue="type",data= year_with_show)
plt.xlabel("")
plt.title('Shows by Country')
plt.ylabel("")
plt.show()
```

```python
# Overall genre makeup of content

plt.figure(figsize=(12,10))
sns.set(style="darkgrid")
ax = sns.countplot(y="listed_in", data=data, palette="Set1", order=data['listed_in'].value_counts().index[0:15])
```

```python
#Duration of TV shows and movies

plt.rcParams['figure.figsize'] = (16,8)
sns.countplot(y = 'duration', data = data, order = data['duration'].value_counts()[:10].index)
plt.title('duration analysis',fontsize = 20);
```
